```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/007ExerciseFile.csv")
```

```python
display(df)
```

```python
hist = sns.displot(df, x="Mutation_rate", hue="Strain", 
            kind="hist", alpha=0.3, bins=55)
plt.xlabel("Mutation Rates")
```

```python
hist = sns.displot(df, x="Mutation_rate", hue="Strain", 
            kind="hist", kde = True, alpha=0.3, bins=55)
```

```python
sns.displot(df, x="Mutation_rate", hue="Strain", 
            kind="kde", fill=True)
```

```python
sns.violinplot(data=df, x="Mutation_rate", y="Strain", 
               width = 1, 
               order = ["st_alpha", "st_beta", "st_delta", "st_gamma"],
              palette = ["green", "red", "orange", "blue"])
```

This violin plot helps show the spread of the detected mutation rates of each the four strain. The alpha strain also seems to have the widest range. 

```python
sns.boxplot(data=df, x="Strain", y="Mutation_rate", 
            order = ["st_alpha", "st_beta", "st_delta", "st_gamma"],
                 palette = ["green", "red", "orange", "blue"])
```

```python
box = sns.boxplot(data=df, x="Strain", y="Mutation_rate", 
            order = ["st_alpha", "st_beta", "st_delta", "st_gamma"],
                 palette = ["green", "red", "orange", "blue"])
box.set_xticklabels(("Alpha","Beta","Delta","Gamma"))

strip = sns.stripplot(data=df, x="Strain", y="Mutation_rate", 
                      alpha = 0.5, 
                      order = ["st_alpha", "st_beta", "st_delta", "st_gamma"],
                     palette = ["green", "red", "orange", "blue"])
strip.set_xticklabels( ("Alpha","Beta","Delta","Gamma"))
box.set_ylabel("Mutation Rates")
strip.set_ylabel("Mutation Rates")
plt.title('Strains and Their Mutation Rates',
         fontsize = 20)
```

### It appears that the Alpha strain has the highest mutation rate by a fairly significant margin. I belive we should be most concerned about this variant, and focus the majority of our efforts to addressing its spread and limiting the its number of mutations.
