```python
#Distribution of the Average for Each Row 
#Y-Axis = Average, X-Axis = Row
plt.plot(df.mean(1))
plt.title('Averages For Each Row',
         fontsize = 20)
plt.xlabel("Row")
plt.ylabel("Average")
```

This appears to be a bimodal distribution as there are two peaks in the data at two different stretches of data (i.e. rows 20-40 and 125-150). 

```python
first_rows = df[0:21,:].std(1)
sec_rows = df[21:42,:].std(1)
```

```python
total_variance = sum((sec_rows*sec_rows)+(first_rows*first_rows))
```

```python
total_variance/40
```

```python
standard_deviation = np.sqrt(0.9777125656582943)
```

```python
plt.imshow(df)
```

This distribution shows off the bimodal distribution through color, with the the two peaks appearing as slightly different in color than the rest of the distribution.

```python
sns.kdeplot(df.mean(1))
plt.title('Density of Each Average',
         fontsize = 20)
plt.xlabel("Averages From the Dataset")
```

This KDE plot shows that there are two means with fairly high density, further providing evidence for a bimodal distribution.

```python
First_mean = df[0:21].mean(1)
Second_mean = df[21:42].mean(1)
```

```python
Smaller_mean = sum(First_mean)
Larger_mean = sum(Second_mean)
```

```python
(Larger_mean - Smaller_mean)/standard_deviation
```
